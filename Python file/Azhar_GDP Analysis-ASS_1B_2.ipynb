##### Categorise the states into four groups based 

```python
# Filter Necessary columns for the State Categorisation

dfPercentile = dfPart_BTranspose
```

```python
# Calculate Quantile for (.2, .5, .85, 1) and make categories

quantile = dfPercentile['Per Capita GSDP (Rs.)'].quantile([.2, .5, .85, 1])

# Add column as Category to represent the state category

# dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'] >= quantile[.85]) & (dfPercentile['Per Capita GSDP (Rs.)'] <= quantile[1]), 'Category'] = 'C1'
# dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'] >= quantile[.5]) & (dfPercentile['Per Capita GSDP (Rs.)'] <= quantile[.85]), 'Category'] = 'C2'
# dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'] >= quantile[.2]) & (dfPercentile['Per Capita GSDP (Rs.)'] <= quantile[.5]), 'Category'] = 'C3'
# dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'] >= 0) & (dfPercentile['Per Capita GSDP (Rs.)'] <= quantile[.2]), 'Category'] = 'C4'

dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'].between(quantile[.85], quantile[1])), 'Category'] = 'C1'
dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'].between(quantile[.5], quantile[.85])), 'Category'] = 'C2'
dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'].between(quantile[.20], quantile[.5])), 'Category'] = 'C3'
dfPercentile.loc[(dfPercentile['Per Capita GSDP (Rs.)'].between(0, quantile[.2])), 'Category'] = 'C4'
```

###### Find the top 3/4/5 sub-sectors

```python
# Filter out Subsectors from the columns

dfSubSectors = dfPercentile.loc[:, dfPercentile.columns.isin(['State','Agriculture, forestry and fishing', 'Mining and quarrying', 'Manufacturing', 'Electricity, gas, water supply & other utility services',
                                                            'Construction', 'Trade, repair, hotels and restaurants', 'Transport, storage, communication & services related to broadcasting','Financial services',
                                                            'Real estate, ownership of dwelling & professional services', 'Public administration','Other services', 'Gross State Domestic Product', 'Per Capita GSDP (Rs.)', 'Category'
                                                           ])]
dfSubSectors
```

```python
# Contribution of Sub Sectors Foreach Category

def contributionOfSubSectors(dfSubSectorsTran):
        
    changeIndex = dfSubSectorsTran.reset_index(drop=True)
    dfSubSectorsTran = changeIndex.T.reset_index()
    dfSubSectorsTran.rename(columns=dict(zip(dfSubSectorsTran.columns[1:], tuple(changeIndex['State']))),inplace=True)
    dfSubSectorsTran = dfSubSectorsTran.rename(columns={'index':'SubSectors'})
    dfSubSectorsTran = dfSubSectorsTran.drop(0, axis=0)
    dfSubSectorsTran = dfSubSectorsTran.reset_index(drop=True)
    dfSubSectorsTran = dfSubSectorsTran.fillna(0)
    dfSubSectorsTran.loc[:,'Total-SubSectors']= dfSubSectorsTran.apply(pd.to_numeric, errors='coerce').sum(axis=1)
    dfSubSectorsTran = dfSubSectorsTran[['SubSectors', 'Total-SubSectors']]
    dfSubSectorsTran['Contribution'] = (dfSubSectorsTran['Total-SubSectors'].values / dfSubSectorsTran[dfSubSectorsTran['SubSectors'] == 'Gross State Domestic Product']['Total-SubSectors'].values) * 100
    dfSubSectorsTran['Cum_percentage'] = dfSubSectorsTran['Contribution'].cumsum().astype(float).round()
    dfSubSectorsTran = dfSubSectorsTran[:-3]
    
    return dfSubSectorsTran


C1 = contributionOfSubSectors(dfSubSectors[dfSubSectors['Category'] == 'C1'])
C2 = contributionOfSubSectors(dfSubSectors[dfSubSectors['Category'] == 'C2'])
C3 = contributionOfSubSectors(dfSubSectors[dfSubSectors['Category'] == 'C3'])
C4 = contributionOfSubSectors(dfSubSectors[dfSubSectors['Category'] == 'C4'])
```

```python
C1 = C1.sort_values(by='Contribution', ascending=False)
C1
```

```python
# Category 1 Graph for subsector contribution

plt.figure(figsize=(15,5))
plt.bar(C1['SubSectors'], C1['Contribution'])
plt.xticks(rotation='90', fontsize='12')
plt.show()
```
